## 3. Model Order Selection 
The data will be split into training and testing using only the attributes of variance, skewness, curtosis and entropy. The training data will have 1000 samples while the testing data will have 372 samples. 

Since the samples are ordered by class, the first half of the dataset is classified as 0 and and the second half of the dataset is classified as 1. The dataset needs to randomized to have 0 and 1 values in both the training and test data. 

```python
#Code taken from various labs from Intro to ML course
cols = df.columns[0:4]
X = np.array(df[cols].values)
Xs = preprocessing.scale(X)
y = np.array(df['Class'])
```

### Logistic Regression
Next, we perform Logistic Regression on the dataset.

1. Create a LogisticRegression object logreg and fit the training data.
2. Measure the accuracy of the classifier by using logreg.predict function to predict labels yhat and measure the fraction of time that the predictions match the true lables.
3. Perform KFold validation with 10 nfolds.

```python
#Create a model
logreg = linear_model.LogisticRegression(C=1e5)
logreg.fit(Xs, y)
```

```python
yhat = logreg.predict(Xs)
acc = np.mean(yhat == y)
print('Accuracy of the training data: {0:f}'.format(acc))
```

#### Interpreting the weight vector (as done in labs for class)
Since the data is somewhat linearly separable, the weight vector can be explicitly retrived and represents the separating hyper-plane between the real and fake banknotes. 

The weight associated to each input dimension (predictor) gives information about its relevance for the discrimination of the real of fake banknotes.

```python
W = logreg.coef_
W = W.ravel()
W
```

Next, we are using KFold Validation to measure the accuracy on the test data with 10 nfolds. The shuffle option is utilized to randomly permutate the test and training sets.

```python
#Use 10 folds for KFold Validation
nfold = 10
kf = KFold(n_splits=nfold,shuffle=True)
prec = []
rec= []
f1 = []
err_rate = []

for Itr, Its in kf.split(Xs):
    #Get training and test data
    Xtr = Xs[Itr,:]
    ytr = y[Itr]
    Xts = Xs[Its, :]
    yts = y[Its]
    
    #Fit a Model
    logreg.fit(Xtr,ytr)
    
    #Predict the labels on the test data
    yhat = logreg.predict(Xts)
    
    #Measure the precisions, recall, and f1-score
    preci,reci,f1i,_= precision_recall_fscore_support(yts,yhat,average='binary')
    prec.append(preci)
    rec.append(reci)
    f1.append(f1i)
    err_rate.append(np.mean(yts != yhat))
    
#Take the mean performance metrics over the different folds
prec = np.mean(prec)
rec=np.mean(rec)
f1=np.mean(f1)
err_mean= np.mean(err_rate)

print('Precision = {0:0.4f}'.format(prec))
print('Recall = {0:0.4f}'.format(rec))
print('F1 = {0:0.4f}'.format(f1))
print('Error Rate = {0:0.4f}'.format(err_mean))
```

### Results:
The accuracy of the training data and the testing data of the model is quite high at 99%.

Even though the accuracy is high, we are going to attempt to model this dataset using a Neural net to see whether the accuracy will change because there might be overfitting of the data using Logistic Regression.

# Neural Net Classifier

We followed the examples from labs and demos to create a neural net model.

```python
#Clear the session
K.clear_session()
```

```python
#Get the training and testing data
ntr = 1000
nts = 372
Xtr, Xts, ytr, yts = train_test_split(X,y,train_size=ntr, test_size=nts,shuffle=True)
```

```python
#Scale the training and test matrices
xmean = np.mean(Xtr, axis = 0)
xstd = np.std(Xtr, axis = 0)
Xtr_scale = (Xtr - xmean[None,:])/xstd[None, :]
Xts_scale = (Xts - xmean[None,:])/xstd[None, :]
```

```python
#Construct the model
nin = Xtr.shape[1]
nout = np.max(ytr) + 1
nh = 256
model = Sequential()
model.add(Dense(nh, input_shape=(nin, ), activation='sigmoid', name= 'hidden'))
model.add(Dense(nout, activation = 'softmax', name = 'output'))
```

```python
#Print the model summary
model.summary()
```

```python
class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        # TODO:  Create two empty lists, self.loss and self.val_acc
        self.loss = []
        self.val_acc = []
    def on_batch_end(self, batch, logs={}):
        # TODO:  This is called at the end of each batch.  
        # Add the loss in logs.get('loss') to the loss list
        self.loss.append(logs.get('loss'))
    def on_epoch_end(self, epoch, logs):
        # TODO:  This is called at the end of each epoch.  
        # Add the test accuracy in logs.get('val_acc') to the val_acc list
        self.val_acc.append(logs.get('val_acc'))
# Create an instance of the history callback
history_cb = LossHistory()
```

```python
opt = optimizers.Adam(lr = 0.001)
model.compile(optimizer=opt, loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
batch_size = 100
model.fit(Xtr_scale, ytr, epochs = 10, batch_size=batch_size, validation_data=(Xts_scale, yts), callbacks=[history_cb])
```

```python
#Plot the accuracy
val_acc = history_cb.val_acc
epochs = len(val_acc)
plt.plot(np.arange(1, epochs+1), val_acc, 'o-', linewidth = 2)
plt.grid()
plt.xlabel('Epoch')
plt.ylabel('Accuracy of testing data')
```

```python
#Loss graph
nsteps = len(history_cb.loss)
ntr = Xtr.shape[0]
epochs = np.arange(1, nsteps + 1)*batch_size/ntr
plt.semilogy(epochs, history_cb.loss)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid()
plt.xlim([0, np.max(epochs)])
plt.tight_layout()
```

### Results:
The accuracy of Neural net classifier is lower than the accuracy with Logistic regression. 

We believe it might be because of the learning rate. Next, we are going to try with a higher learning rate of 0.01. 

```python
#Clear the session
K.clear_session()

#Construct the model
model = Sequential()
model.add(Dense(nh, input_shape=(nin, ), activation='sigmoid', name= 'hidden'))
model.add(Dense(nout, activation = 'softmax', name = 'output'))

#Optimizer
opt = optimizers.Adam(lr = 0.01)
model.compile(optimizer=opt, loss ='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
batch_size = 100
model.fit(Xtr_scale, ytr, epochs = 10, batch_size=batch_size, validation_data=(Xts_scale, yts), callbacks=[history_cb])
```

```python
#Plot the accuracy
val_acc = history_cb.val_acc
epochs = len(val_acc)
plt.plot(np.arange(1, epochs+1), val_acc, 'o-', linewidth = 2)
plt.grid()
plt.xlabel('Epoch')
plt.ylabel('Accuracy of testing data')
```

```python
#Loss graph
nsteps = len(history_cb.loss)
ntr = Xtr.shape[0]
epochs = np.arange(1, nsteps + 1)*batch_size/ntr
plt.semilogy(epochs, history_cb.loss)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid()
plt.xlim([0, np.max(epochs)])
plt.tight_layout()

#Results: this loss function converges faster than the one because the learning rate is higher.
#However, the accuracy of this learning is higher and more along the lines of Logistic Regression.
```

## Conclusion:

With the learning rate of 0.01, the Neural net classifier gives an accuracy of 98% which is very close to the accuracy of the Logistic Regression (if you think about the precision of the testing data).

#### So, why did the learning rate have to increased?
The learning rate is used to find the global minima. If the learning rate is small, the model reaches the global minima with better accuracy because it is slow. However, if it is too slow, we might end up reaching a local minima rather than the global minima (also depending on where you begin traversal). For instance, before we tried the learning rate of 0.01, we also tried 0.001, but the accuracy then was around 68% whch is quite low.

Therefore, we concluded that the learning rate of 0.01 is the optimal learning rate. 

Furthermore, for this dataset, Logistic Regression is sufficient enough to provide accurate results, and a neural net is not necessary. However, it can help with understanding the dataset better.
